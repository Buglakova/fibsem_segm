```python
import numpy as np
import matplotlib.pyplot as plt


from skimage.transform import downscale_local_mean, rescale
from skimage.morphology import cube, square
from skimage.filters.rank import minimum
```

```python
x = np.zeros((20, 20))
x[5:15, 5:15] = 1
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(x, interpolation="nearest")
plt.colorbar()
```

```python
x_min = minimum(x, square(2))
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(x_min, interpolation="nearest")
plt.colorbar()
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(x_min[::2, ::2], interpolation="nearest")
plt.colorbar()
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(x_min - x, interpolation="nearest")
plt.colorbar()
```

```python
x_mean = downscale_local_mean(x, (2, 2))
plt.figure(figsize=(8, 8))
plt.imshow(x_mean, interpolation="nearest")
plt.colorbar()
```

```python
x_mean = downscale_local_mean(x, (2, 2))
plt.figure(figsize=(8, 8))
plt.imshow(x_mean, interpolation="nearest")
plt.colorbar()
```

```python
x_rescale = rescale(x, 0.5, anti_aliasing=False, order=0)
plt.figure(figsize=(8, 8))
plt.imshow(x_rescale, interpolation="nearest")
plt.colorbar()
```

```python
x_rescale = rescale(x, 0.5, anti_aliasing=True)
plt.figure(figsize=(8, 8))
plt.imshow(x_rescale, interpolation="nearest")
plt.colorbar()
```
